```python
# Segmentación de Clientes de E-commerce mediante Técnicas de Clustering

## Objetivo

El objetivo de este proyecto es identificar grupos de clientes con comportamientos similares utilizando técnicas de aprendizaje no supervisado. La segmentación permitirá comprender mejor los patrones de compra y generar estrategias de marketing diferenciadas para cada grupo de clientes.

---

# Carga y Exploración Inicial de los Datos

Se carga el conjunto de datos que contiene información demográfica y de comportamiento de clientes de una plataforma de comercio electrónico.

Las variables incluyen:

* Edad del cliente.
* Género.
* Frecuencia mensual de compras.
* Gasto promedio por compra.
* Tiempo promedio de navegación.
* Uso de dispositivos móviles.
* Categorías preferidas.
* Respuesta a promociones.
* Calificación del servicio.
* País de residencia.

Posteriormente se inspecciona la estructura general del dataset mediante:

* Visualización de registros.
* Número de filas y columnas.
* Tipos de datos.
* Identificación de variables cuantitativas y cualitativas.

---

# Limpieza y Preparación de los Datos

## Tratamiento de Valores Faltantes

Los valores faltantes en variables numéricas fueron imputados utilizando la media de cada variable.

Esta estrategia permite conservar los registros sin introducir sesgos importantes en la distribución de los datos.

## Eliminación de Duplicados

Se verificó la existencia de registros duplicados y se eliminaron aquellos que aparecían repetidos para evitar distorsiones en los análisis posteriores.

---

# Análisis Exploratorio de Datos

## Estadística Descriptiva

Se calcularon medidas descriptivas para las variables numéricas:

* Media.
* Mediana.
* Desviación estándar.
* Valores mínimos y máximos.
* Cuartiles.

Estas métricas permiten comprender la distribución y dispersión de los datos.

## Análisis de Correlación

Se calculó la matriz de correlación para identificar relaciones entre variables numéricas.

Posteriormente se construyó un mapa de calor (Heatmap) para facilitar la interpretación visual de las asociaciones existentes entre variables.

## Distribución de Variables

Se utilizaron histogramas y gráficos de densidad para evaluar la forma de las distribuciones de las variables numéricas.

Asimismo, se aplicó la regla de Sturges para determinar un número adecuado de intervalos en los histogramas.

---

# Detección y Tratamiento de Valores Atípicos

Se identificaron valores atípicos utilizando el método del rango intercuartílico (IQR).

Los registros que se encontraban fuera de los límites establecidos fueron eliminados para reducir el impacto de observaciones extremas sobre los algoritmos de clustering.

Esta etapa contribuye a mejorar la estabilidad y calidad de los agrupamientos obtenidos.

---

# Preprocesamiento de Datos

## Transformación de Variables Numéricas

Las variables numéricas fueron escaladas mediante RobustScaler.

Este escalador es menos sensible a valores extremos que StandardScaler, por lo que resulta apropiado para este tipo de datos.

## Transformación de Variables Categóricas

Las variables categóricas nominales fueron codificadas mediante OneHotEncoder.

Las variables binarias fueron transformadas mediante técnicas de codificación apropiadas para preservar la información y permitir su utilización por los algoritmos de clustering.

## Construcción del Pipeline

Se implementó un ColumnTransformer para aplicar automáticamente las transformaciones necesarias a cada grupo de variables.

Esto garantiza un flujo de procesamiento reproducible y consistente.

---

# Determinación del Número de Clusters

## Método del Codo

Se aplicó el método del codo para analizar la evolución de la inercia al incrementar el número de clusters.

La inercia mide la compactación interna de los grupos.

## Coeficiente de Silhouette

Posteriormente se calculó el coeficiente de Silhouette para distintos valores de K.

El mejor resultado se obtuvo para:

**K = 2**

Por lo tanto, se seleccionó este valor como número óptimo de clusters.

---

# Clustering mediante K-Means

Se entrenó un modelo K-Means utilizando dos clusters.

El algoritmo busca minimizar la distancia entre cada observación y el centroide de su grupo correspondiente.

Los centroides obtenidos representan el perfil promedio de cada segmento de clientes.

## Interpretación de Centroides

Se utilizaron gráficos de barras y gráficos radar para visualizar las características predominantes de cada cluster.

Estas representaciones permiten identificar diferencias significativas entre los grupos encontrados.

---

# Análisis de Componentes Principales (PCA)

Con el fin de reducir la dimensionalidad y facilitar la visualización de los resultados, se aplicó Análisis de Componentes Principales (PCA).

## Objetivos del PCA

* Reducir la complejidad de los datos.
* Eliminar redundancias.
* Facilitar la visualización de clusters.
* Evaluar la varianza explicada por cada componente.

## Varianza Explicada

Se analizó la contribución de cada componente principal para determinar cuánta información del conjunto original es retenida tras la transformación.

## Visualización de Clusters

Los clientes fueron proyectados sobre los dos primeros componentes principales para observar la separación entre grupos.

---

# Clustering Jerárquico

Además de K-Means, se evaluaron distintos métodos de agrupamiento jerárquico:

* Ward.
* Average.
* Single.
* Complete.

## Dendrogramas

Se construyeron dendrogramas para analizar visualmente el proceso de agrupamiento.

Estos diagramas permiten observar la distancia a la que se fusionan los clusters y ayudan a determinar un número adecuado de grupos.

## Selección de Clusters

A partir de los dendrogramas se realizaron cortes sobre el árbol jerárquico para generar agrupamientos comparables con los obtenidos mediante K-Means.

---

# Validación de la Estructura de Clusters

## Estadístico de Hopkins

Se utilizó el estadístico de Hopkins para evaluar si los datos presentan una estructura natural de agrupamiento.

Un valor cercano a 1 indica alta tendencia a formar clusters.

El valor obtenido fue aproximadamente:

**Hopkins = 0.63**

Este resultado indica una tendencia moderada a la formación de grupos.

---

# Comparación de Algoritmos

Se compararon distintos algoritmos utilizando métricas internas de validación:

## Silhouette Score

Evalúa la separación entre clusters.

* Valores cercanos a 1 indican mejor agrupamiento.
* Valores cercanos a 0 indican solapamiento.

## Calinski-Harabasz

Mide la relación entre la dispersión entre clusters y la dispersión interna.

Valores mayores indican mejor separación.

## Davies-Bouldin

Evalúa la similitud entre clusters.

Valores menores indican mejor calidad de agrupamiento.

Los resultados mostraron que los métodos jerárquicos Average y Single obtuvieron los mejores indicadores de separación, mientras que K-Means presentó un comportamiento competitivo y más fácil de interpretar.

---

# Comparación entre PCA y Datos Originales

Se evaluó el desempeño de K-Means utilizando:

1. Datos preprocesados originales.
2. Datos reducidos mediante PCA.

Los resultados mostraron una mejora en la calidad de la segmentación al utilizar PCA, evidenciada por un incremento en el coeficiente de Silhouette y una reducción del índice Davies-Bouldin.

---

# Interpretación de los Clusters

## Cluster 1

Clientes con mayor interacción con la plataforma:

* Mayor frecuencia de compra.
* Mayor gasto promedio.
* Mayor tiempo de navegación.
* Mayor respuesta a promociones.

Estos clientes representan el segmento de mayor valor para la empresa.

## Cluster 2

Clientes con menor actividad:

* Menor frecuencia de compra.
* Menor gasto promedio.
* Menor participación en promociones.
* Menor interacción general con la plataforma.

Este grupo representa clientes ocasionales o de baja fidelización.

---

# Conclusiones

Los resultados obtenidos muestran que los clientes pueden segmentarse efectivamente en dos grupos principales.

El análisis de Silhouette indicó que dos clusters constituyen la mejor partición para los datos estudiados.

La aplicación de PCA mejoró significativamente la calidad de la segmentación al incrementar la separación entre grupos.

La comparación de algoritmos mostró que los métodos jerárquicos Average y Single presentaron los mejores indicadores de agrupamiento; sin embargo, K-Means fue seleccionado como modelo principal debido a su simplicidad, estabilidad y facilidad de interpretación.

La segmentación obtenida proporciona información útil para el diseño de estrategias comerciales diferenciadas, campañas de marketing personalizadas y programas de fidelización dirigidos a cada perfil de cliente.
```
